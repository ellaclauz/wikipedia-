```python
import pandas as pd

wikipedia = pd.read_excel('wikipedia_dataset_flat.xlsx')


#feature extraction 
def feature_extraction(wikipedia):
    wikipedia['Agents'] == wikipedia('page').apply(lambda x: x.split('_')[-1])
    wikipedia['Access'] == wikipedia('page').apply(lambda x: x.split('_')[-2])
    wikipedia['Temp'] == wikipedia('page').apply(lambda x: x.split('_')[-3])
    wikipedia['Language'] == wikipedia('page').apply(lambda x: x.split('_')[0])
    wikipedia = wikipedia.drop(columns= ['Temp'])
    return(wikipedia)

wikipedia_final = feature_extraction

wikipedia_final
```
